```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

a = pd.read_csv("adult.csv")
df_adult = pd.DataFrame(a)

df_adult.replace("?", pd.NA, inplace=True)

df_adult['workclass'] = df_adult['workclass'].fillna(df_adult['workclass'].mode()[0])
df_adult['occupation'] = df_adult['occupation'].fillna(df_adult['occupation'].mode()[0])
df_adult['native-country'] = df_adult['native-country'].fillna(df_adult['native-country'].mode()[0])

ordinal_encoder = OrdinalEncoder(categories=[['Male', 'Female']])
df_adult['gender_encoded'] = ordinal_encoder.fit_transform(df_adult[['gender']])

onehot_encoder = OneHotEncoder(sparse_output=False)  # Use sparse_output=False instead of sparse=False

encoded_workclass = onehot_encoder.fit_transform(df_adult[['workclass']])
encoded_workclass_df = pd.DataFrame(encoded_workclass, columns=onehot_encoder.get_feature_names_out(['workclass']))

encoded_occupation = onehot_encoder.fit_transform(df_adult[['occupation']])
encoded_occupation_df = pd.DataFrame(encoded_occupation, columns=onehot_encoder.get_feature_names_out(['occupation']))

encoded_income = onehot_encoder.fit_transform(df_adult[['income']])
encoded_income_df = pd.DataFrame(encoded_income, columns=onehot_encoder.get_feature_names_out(['income']))

df_encoded = pd.concat([df_adult, encoded_workclass_df, encoded_occupation_df, encoded_income_df], axis=1)

df_encoded.drop(['gender', 'workclass', 'occupation', 'income'], axis=1, inplace=True)

print(df_encoded.head())
```
